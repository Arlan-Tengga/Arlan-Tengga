```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,2*np.pi,1000)
r = 1 * (3/4)*np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x,y)
```
